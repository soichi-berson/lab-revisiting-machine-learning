### Identify columns that over 85% missing values and add them to the previous list.

```python
# create the function which creates the list
def drop_columns (df: pd.DataFrame, threshold=0.25)->list:
    '''
    This function takes a dataframe and a float as an input and returns a 
    list of columns with null vales greater than the specified threshold.
    
    Input: 
    df:pd.DataFrame
    threshhold: float
    
    Output:list 
    '''
    nulls_percent_df = pd.DataFrame(df.isna().sum()/len(df)).reset_index()
    nulls_percent_df.columns = ['column_name', 'nulls_percentage']
    columns_above_threshold = nulls_percent_df[nulls_percent_df['nulls_percentage']>threshold]
    drop_columns_list = list(columns_above_threshold['column_name'])

    return drop_columns_list
```

```python
missing_values_list = drop_columns(data,0.85)
missing_values_list
```

```python
# append the list
for column in missing_values_list:
    drop_list.append(column)

drop_list   
```

### Remove the columns included in the drop_list from the dataframe

```python
data = data.drop(columns=drop_list)
data.shape
```

### Now, reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

#### Note that there are a few null values in the column. We will first replace those null values using the code below:
##### print(categorical['GENDER'].value_counts())
##### categorical['GENDER'] = categorical['GENDER'].fillna('F')

```python
numerical = data.select_dtypes(np.number)
categorical = data.select_dtypes([object])
```

```python
categorical.shape
```

```python
categorical['GENDER'].value_counts()
```

```python
categorical['GENDER'].unique()
```

```python
# replace ' ' with np.NaN
categorical['GENDER'] = categorical['GENDER'].replace(" ", np.NaN)
categorical['GENDER'].unique()
```

```python
### create the function of others 

def clean_gender(x):
    other_gender = [ np.nan, 'C', 'U', 'J', 'A']
    if x in other_gender:
        return 'other'
    else:
        return x

categorical['GENDER'] = list(map(clean_gender, categorical['GENDER']))
categorical['GENDER'].unique()
```
